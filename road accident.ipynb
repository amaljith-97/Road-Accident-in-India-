## data visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize=[20,8])
print ('countplot for Region')
sns.countplot(df['region'],data = df,palette = 'rocket')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=[20,8])
print('pieplot for Region')
df['region'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.xticks(rotation=0)
plt.show()
```

```python
df['regionid'].unique()
```

```python
df['regionid'].value_counts()
```

```python
plt.figure(figsize=[20,8])
print('countplot of RegionID')
sns.countplot(df['regionid'],data = df, palette='mako')
plt.xticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize=[20,8])
print ('pieplot for RegionID')
df['regionid'].value_counts().plot(kind='pie',autopct= '%1.1f%%')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=[20,8])
print('Pairplot')
sns.pairplot(df,hue='region')
plt.xticks(rotation=0)
plt.show()
```

```python
import numpy as np
```

```python
plt.figure(figsize=[20,12],dpi=250)
plt.title('correlation graph', fontsize = 14)
matrix = np.triu(df.corr())
sns.heatmap(df.corr(),annot=True,cmap = 'coolwarm', cbar = False)
plt.show()
```
